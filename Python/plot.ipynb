```python
"""
Spannungsdaten aus CSV einlesen und plotten
Autor: ChatGPT
"""

import pandas as pd
import matplotlib.pyplot as plt

# ========================== #
# ==== KONFIGURATION ==== #
# ========================== #
CSV_FILE = "/Users/Ben/Desktop/Mikrowellen/Data/data_2025_11_21_0_58Degree.csv"  # Pfad zur CSV-Datei
#CSV_FILE = "/Users/Ben/Desktop/Mikrowellen/Data/data-2025-11-03-T14-49-10.csv"
COLUMN_NAME = "Voltage (V)"  # Spaltenname mit den Spannungswerten

# ========================== #
# ==== CSV einlesen ==== #
# ========================== #
try:
    df = pd.read_csv(CSV_FILE)
except FileNotFoundError:
    print(f"Datei {CSV_FILE} nicht gefunden!")
    exit(1)

if COLUMN_NAME not in df.columns:
    print(f"Spalte '{COLUMN_NAME}' nicht gefunden! Verfügbare Spalten: {df.columns.tolist()}")
    exit(1)

voltages = df[COLUMN_NAME]

# Zeitachse (angenommen 1 Sample pro Einheit, kann angepasst werden)
time = range(len(voltages))

# ========================== #
# ==== Plotten ==== #
# ========================== #
plt.figure(figsize=(10, 5))
plt.plot(time, voltages, label="Spannung")
plt.xlabel("Sample")
plt.ylabel("Spannung (V)")
plt.title("Spannungsdaten aus CSV")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
