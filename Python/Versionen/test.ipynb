```python
import serial
import time
import numpy as np
```

```python
print("start")
ser = serial.Serial("/dev/cu.usbmodem141101", 115200, timeout=1)
time.sleep(1)

def send(cmd):
    ser.write((cmd + "\r").encode())
    time.sleep(0.1)
    response = ser.read_all().decode().strip()
    return response
```
